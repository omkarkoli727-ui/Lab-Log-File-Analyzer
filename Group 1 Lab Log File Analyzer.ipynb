```python
from collections import Counter, defaultdict

filename = "D:\Group 1 mini project\Group 1 Activity_log4.txt"

student_time = defaultdict(int)
activity_counter = Counter()
lab_time_per_day = defaultdict(int)

with open(filename, "r") as file:
    for line in file:
        line = line.strip()

        # Skip empty lines
        if not line:
            continue

        parts = [p.strip() for p in line.split(",")]

        # Skip header or bad rows
        if parts[3].lower() == "duration":
            continue

        if len(parts) != 4:
            continue

        date, student, activity, duration = parts

        try:
            duration = int(duration)
        except ValueError:
            continue

        student_time[student] += duration
        activity_counter[activity] += 1

        if activity.lower() == "lab":
            lab_time_per_day[date] += duration

# Find most common activity
most_common_activity, activity_count = activity_counter.most_common(1)[0]

# -------- Summary Report --------
print("\n===== STUDENT ACTIVITY SUMMARY REPORT =====\n")

print("Total Time Spent by Each Student:")
for student, time in student_time.items():
    print(f"  {student}: {time} minutes")

print("\nMost Common Activity:")
print(f"  {most_common_activity} ({activity_count} occurrences)")

print("\nTotal Time Spent in Labs Per Day:")
for date, time in lab_time_per_day.items():
    print(f"  {date}: {time} minutes")

print("\n===========================================\n")
```
